```python
class Stack:
    def __init__(self):
        self.stack = []

    def insert(self, value):
        self.stack.append(value)
        print(f"Inserted {value} into stack.")

    def delete(self):
        if not self.is_empty():
            value = self.stack.pop()
            print(f"Deleted {value} from stack.")
        else:
            print("Stack is empty.")

    def display(self):
        if self.is_empty():
            print("Stack is empty.")
        else:
            print("Stack contents:", self.stack)

    def is_empty(self):
        return len(self.stack) == 0


class Queue:
    def __init__(self):
        self.queue = []

    def insert(self, value):
        self.queue.append(value)
        print(f"Inserted {value} into queue.")

    def delete(self):
        if not self.is_empty():
            value = self.queue.pop(0)
            print(f"Deleted {value} from queue.")
        else:
            print("Queue is empty.")

    def display(self):
        if self.is_empty():
            print("Queue is empty.")
        else:
            print("Queue contents:", self.queue)

    def is_empty(self):
        return len(self.queue) == 0


def main():
    while True:
        print("\nChoose an option:")
        print("1. Stack")
        print("2. Queue")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            stack = Stack()
            while True:
                print("\nStack Operations:")
                print("1. Insert")
                print("2. Delete")
                print("3. Display")
                print("4. Back to Main Menu")
                sub_choice = input("Enter your choice: ")

                if sub_choice == '1':
                    value = input("Enter value to insert: ")
                    stack.insert(value)
                elif sub_choice == '2':
                    stack.delete()
                elif sub_choice == '3':
                    stack.display()
                elif sub_choice == '4':
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == '2':
            queue = Queue()
            while True:
                print("\nQueue Operations:")
                print("1. Insert")
                print("2. Delete")
                print("3. Display")
                print("4. Back to Main Menu")
                sub_choice = input("Enter your choice: ")

                if sub_choice == '1':
                    value = input("Enter value to insert: ")
                    queue.insert(value)
                elif sub_choice == '2':
                    queue.delete()
                elif sub_choice == '3':
                    queue.display()
                elif sub_choice == '4':
                    break
                else:
                    print("Invalid choice. Please try again.")

        elif choice == '3':
            print("Exiting...")
            break
        else:
            print("Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```
